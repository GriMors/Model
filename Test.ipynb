# Импорт модулей и загрузка файлов

```python
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from google.colab import files
uploaded = files.upload()
```

### **Работа над файлом "model_data_cults_100i-50c-50t.csv".**
Количество Культов.
Начальные условия: *100 повторов, 50 культов, 50 дней*.

```python
data = pd.read_csv(io.StringIO(uploaded['model_data_cults_100i-50c-50t.csv'].decode('utf-8')))
data
```

```python
n = len(data.axes[0])
x = []
y = []
i = 0

for data.axes[1] in data:
  i += 1
  x.append(i-1)
  y.append(int(data[f"{i-1}"].sum()/n))

print(y)
print(x)
```

```python
plt.plot(x, y, label="line С")
plt.plot()

plt.xlabel("Дни")
plt.ylabel("Культы")
plt.title("Изменение количества культов")
plt.legend()
plt.show()
```
